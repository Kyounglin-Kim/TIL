```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. 각 파일을 pd.read_csv() 로 읽어들인다
# 2. 데이터 확인 및 구조분석과 각 데이터 프레임에 헤더를 추가
ratings = pd.read_csv('./data/ratings.dat',sep='::',names=['userId','movieId','rating','timestamp'],header=None)
```

```python
ratings.head()
```

```python
movies = pd.read_csv('./data/movies.dat',sep='::',names=['movieId','title','genres'],header=None)
```

```python
movies.head()
```

```python
users = pd.read_csv('./data/users.dat',sep='::',header=None)
```

```python
users.head()
```

```python
# 열 삭제
# users_df.drop([3,4],axis=1,inplace=True)
users=users.iloc[:,0:3]
```

```python
users.columns=['userId','gender','age']
```

```python
users.head()
```

```python
# 3. 공통 컬럼을 이용한 데이터 병합 ratings/movies/users
# ra_mo=pd.merge(ratings,movies,on='movieId',how='inner')
# display(ra_mo.head())
# data=pd.merge(ra_mo,users,on='userId',how='inner')
# display(data.head())

data = pd.merge(pd.merge(ratings,users),movies)
data.head()
```

```python
# 4. user 평점 갯수 순위 확인
four = data.pivot_table('rating','userId',aggfunc='count')
four['rank']=four.rank(ascending=False)
four=four.sort_values(by='rating',ascending=False)
four.head(10)

```

```python
# 5. 각 영화의 성별 평균 평점

five=data.pivot_table('rating','title','gender',aggfunc='mean')
five.head(10)


```

```python
# 6. 평점이 220개 이상인 영화만 필터링

six = data.pivot_table('movieId','title',aggfunc='count')
six = six[six['movieId']>=220]
six.head(10)


```

```python
# 6번 강사님 답
rating_by_title = data.groupby('title').size()
rating220 = rating_by_title.index[rating_by_title>=220]
rating220
```

```python
# 7. 평점이 200개 이상인 영화들의 성별 평균 평점 - 풀이 1

m_cnt = pd.DataFrame({'m_cnt' : data['movieId'].value_counts()})
m_cnt.head()
```

```python
m_cnt=m_cnt.reset_index()
m_cnt.head()
```

```python
m_cnt.columns=['movieId','m_cnt']
```

```python
result_df = pd.merge(data,m_cnt,on='movieId',how='left')
result_df.head()
```

```python
result_df=result_df[result_df['m_cnt']>=200]
```

```python
seven=result_df['rating'].groupby([result_df['title'],result_df['gender']]).mean()
seven.head()
```

```python
# 7. 평점이 200개 이상인 영화들의 성별 평균 평점 - 풀이 2
seven = data.pivot_table('movieId','title',aggfunc='count')
seven = seven[seven['movieId']>=200]
seven

re_seven = pd.merge(seven,data,on='movieId',how='left')
re_seven.pivot_table('rating','title','gender',aggfunc='mean').head()
```

```python
seven = data.pivot_table('title','movieId',aggfunc='count')
seven = seven[seven['title']>=200]
# seven

re_seven = pd.merge(seven,data,on='movieId',how='left')
re_seven
re_seven.pivot_table('rating','title_y','gender',aggfunc='mean').head()
```

```python
# 강사님 풀이
rating_mean_df = data.pivot_table('rating','title','gender',aggfunc='mean')
rating_by_title = data.groupby('title').size()
rating200 = rating_by_title.index[rating_by_title>=200]
rating200_mean = rating_mean_df.loc[rating200]
rating200_mean.head()
```

```python
# 8. 여성이 가장 좋아하는 영화 순위
```

```python
f_df=data[data['gender']=='F']
eight=pd.DataFrame(f_df.groupby(f_df['title']).size())
rank= eight.rank(ascending=False)
eight = pd.concat([eight,rank],axis=1)
eight.columns=['count','rank']
eight
eight.sort_values(by='rank',ascending=True).head()
```

```python
# 9. 남녀간의 평균 평점 차이

nine=data['rating'].groupby(data['gender']).mean()
nine=pd.DataFrame(nine)
nine.loc['F']-nine.loc['M']
```

```python
# 10. 평점의 표준편차가 큰 영화

# ten=data['rating'].groupby(data['title']).std()
# ten=pd.DataFrame()

ten = data.pivot_table('rating','title',aggfunc='std')
ten = ten.sort_values(by='rating',ascending=False)
ten.head(1)
```

```python
# 추가적인 요구사항 1) Comedy 영화 중 가장 평점이 낮은 영화의 제목
plus1=data[data['genres']=='Comedy']
plus1=plus1.pivot_table('rating','title',aggfunc='mean',fill_value=0)
plus1=plus1.sort_values(by='rating',ascending=True)
plus1.head(6)
```

```python
# 추가적인 요구사항 2) 평균 평점이 가장 높은 영화의 제목(동률이 있을경우 모두 출력)
plus2=data.pivot_table('rating','title',aggfunc='mean').sort_values(by='rating',ascending=False)
plus2.head(10)
```
