```python
import folium as g
```

```python
# 37.509087, 127.055704
g_map = g.Map(location=[37.509087, 127.055704],zoom_start=18,
             tiles = 'Stamen Terrain' # 지도의 스타일 변화 - Stamen Toner도 있음 )

# option
marker01 = g.Marker([37.509087, 127.055704],popup='campus seven'
                  ,icon = g.Icon(color='blue'))

marker01.add_to(g_map)

marker02 = g.CircleMarker([37.509087, 127.055704],
                         radius=100, # 반경을 잡아줄수 있음 
                         color='skyblue',
                         popup ='campus seven',
                         fill_color = 'skyblue') 

marker02.add_to(g_map)

g_map # 틀과 옵셥을 만들고 마지막에 실행

#g_map.save('./data/now_location_map.html')
```

```python
# 파일읽기 : 서울지역 대학교 위치 정보들

seoul_uni_df=pd.read_excel('./data/서울지역 대학교 위치.xlsx')
seoul_uni_df.head()
```

```python
seoul_map = g.Map(location=[37.55,126.98],
                 tiles='Stamen Terrain',
                 zoom_start=12)

#seoul_uni_df.shape[0] # 이 수만큼 루프 돌려야함
for idx in range(len(seoul_uni_df)):
    marker01 = g.Marker(seoul_uni_df.iloc[idx, [0,1]],
                        popup=seoul_uni_df.index[idx],
                  icon = g.Icon(color='blue'))

    marker01.add_to(seoul_map)

seoul_map
```

```python
seoul_map = g.Map(location=[37.55,126.98],
                 tiles='Stamen Terrain',
                 zoom_start=12)

for name, lat, lng in zip(seoul_uni_df.index, seoul_uni_df.위도, seoul_uni_df.경도):
    g.Marker([lat,lng],popup=name, icon=g.Icon(color='red')).add_to(seoul_map)

    
    g.CircleMarker([lat,lng],popup=name, icon=g.Icon(color='red'),
            radius=20,
            fill=True,
            color='red',
            fill_color='coral',
            fill_opacity=0.3).add_to(seoul_map)
seoul_map
```

```python
# pd.Excelpile?은 시트가 여러개 일때 사용
elem_school=pd.read_excel('./data/elem_school_folium.xlsx')
elem_school.head()
```

```python
# Feature Selection
# 학교명, 학교급구분, 소재지도로명, 위도, 경도
elem_school_df=elem_school.filter(['학교명','학교급구분','소재지도로명주소','위도','경도'])
elem_school_df.head()
```

```python
# 소재지도로명주소가 서울을 포함하고 학교급구분 초등학교만 추출
elementary=elem_school_df[(elem_school_df['학교급구분']=='초등학교')&(elem_school_df['소재지도로명주소'].str[:5]=='서울특별시')]
elementary.head()

# str.contains('서울')로 찾아도 된다
```

```python
elementary_map = g.Map(location=[37.55,126.98],
                 zoom_start=12)

for name, lat, lng in zip(elementary.학교명, elementary.위도, elementary.경도):
    g.Marker([lat,lng],popup=name, icon=g.Icon(color='purple')).add_to(elementary_map)

    
#     g.CircleMarker([lat,lng],popup=name, icon=g.Icon(color='red'),
#             radius=20,
#             fill=True,
#             color='red',
#             fill_color='coral',
#             fill_opacity=0.3).add_to(elementary_map)
elementary_map
```

```python
# 경기도인구데이터 불러오기
gyeonggi=pd.read_excel('./data/경기도인구데이터.xlsx',index_col='구분')
gyeonggi.head()
```

```python
# 경계정보를 가진 json 파일 읽어보기
geo_data = json.load(open('./data/경기도행정구역경계.json',encoding='utf-8'))
```

```python
gyeonggi.columns=gyeonggi.columns.map(str)

# 지도 생성
gyeonggi_map = g.Map(location=[37.55,126.98],
                 zoom_start=12)

# 구분 경계 표시할때 사용하는 객체
year='2017'
g.Choropleth(geo_data=geo_data,
            data = gyeonggi[year],
            columns =  [gyeonggi.index,gyeonggi[year]],
            key_on='feature.properties.name').add_to(gyeonggi_map )

gyeonggi_map 
```
