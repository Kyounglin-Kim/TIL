- fillna : NaN을 값을 원하는 값으로 변경할 수 있다

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_df = pd.DataFrame({
    'A' : [1,2,3,4,3],
    'B' : [2,3,1,2,2],
    'C' : [1,5,2,4,4]
})

sample_df
```

```python
sample_df['C'][2] = np.nan # sample_df.iloc[2,2] = np.nan 도 가능
```

```python
# na가 여러개 일때 저렇게 결측치를 변환하여 처리
sample_df.apply(pd.value_counts).fillna(0.0)
```

- astype : 자료형을 바꾸는 기능

```python
# 0을 int로 자료 바꾸기
sample_df.apply(pd.value_counts).fillna(0).astype(int)
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
# 타이타닉 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이 값으로 대체
titanic['age']=titanic['age'].fillna(np.mean(titanic['age'])).astype(float)
print(titanic['age'].fillna(np.mean(titanic['age'])).astype(float).head())
# 타이타닉 승객에 대해 나이와 성별에 의한 age_gender_cat 열을 만들고 

# 조건 1 : 성별을 나탄내는 문자를 male 또는 female로 시작한다

# 조건 2 : 성별을 나타내는 문자열 뒤에 나이를 나타내는 문자열로 변경하라

# 조건 3 : 예시) 남성의 나이가 27이라면 -> male27

titanic['age_gender_cat']=titanic['sex']+titanic['age'].astype(int).apply(str)
display(titanic['age_gender_cat'].head(10))
```

```python
# 5명의 학생의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 만든다.


# 1. 
# 학생 이름을 나타내는 열을 포함시키지 않고 
# 데이터프레임 df_score1 을 생성한 후, 
# df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다. 
# reset_index 명령으로 이 인덱스 열을 명령으로 
# 일반 데이터열로 바꾸여 데이터프레임 df_score2을 만든다.


df_score1 = pd.DataFrame({
    '국어' : [50,90,95,60,80],
    '영어' : [60,95,85,75,80],
    '수학' : [70,75,90,65,80],
})

df_score1.index = ['짱구','철수','맹구','훈이','유리']
df_score1.index.name = '이름'
display(df_score1)
df_score2=df_score1.reset_index()
display(df_score2)


# 2.
# 학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 df_score2에 
# set_index 명령을 적용하여 다시 학생 이름을 나타내는 열을 인덱스로 변경한다.

df_score2 = df_score2.set_index('이름')
display(df_score2)



```

### DataFrame merge

```python
# DataFrame merge
data1 = {
    '학번' : [1, 2, 3, 4],
    '이름' : ['섭섭해', '김한준', '김선림', '최호진'],
    '학년' : [2, 4, 1, 3]
}

data2 = {
    '학번' : [1, 2, 4, 5],
    '학과' : ['CS', 'MATH', 'MATH', 'CS'],
    '학점' : [2.4, 4.5, 3.4, 4.2]
}
```

```python
stu_df = pd.DataFrame(data1)
major_df = pd.DataFrame(data2)
display(stu_df)
display(major_df)
```

```python
# 학번을 기준으로 merge가 됨 - 매칭 되지 않는것은 누락이 된다
pd.merge(stu_df,major_df)
```

```python
# on -> 셀을 병합할 열의 기준 how -> merge 종류
pd.merge(stu_df,major_df,on='학번',how='inner') # default
```

```python
pd.merge(stu_df,major_df,on='학번',how='left') # 왼쪽기준
```

```python
pd.merge(stu_df,major_df,on='학번',how='right') # 오른쪽기준
```

```python
pd.merge(stu_df,major_df,on='학번',how='outer') # 전체 
```

- 컬럼인덱스가 아닌 인덱스를 기준으로 병합 
- left_index , right_index

```python
pop_df01 = pd.DataFrame({
    'city' : ['seoul','seoul','seoul','busan','busan'],
    'year' : [2010, 2005, 2020, 2018,2015],
    'pop' : [1234567,2345678,3456789,4567890,5678901]
})

pop_df01
```

```python
pop_df02    = pd.DataFrame( np.arange(12).reshape((6,2)) , 
    index   = [['busan' , 'busan', 'seoul' , 'seoul' , 'seoul' , 'seoul'] , 
               [2010, 2005, 2020, 2018, 2015, 2010]] , 
    columns =  ['col01','col02']
)

pop_df02
```

```python
pd.merge(pop_df01,pop_df02,left_on=['city','year'],right_index=True)
```

- 다중 인덱스가 아닌 단일 인덱스를 이용하는 방법 

```python
data1 = { "이름" : ["이지안","박동훈","이순신","강감찬"],
          "학년" : [2,4,1,3]}

data2 = { "학과" : ["CS","MATH","MATH","CS"],
          "학점" : [3.4,2.9,4.5,1.2]}
```

```python
df01 = pd.DataFrame(data1, index=[1,2,3,4])
df02 = pd.DataFrame(data2, index=[1,2,4,5])
```

```python
display(df01)
display(df02)
```

```python
# right_index=True,left_index=True -> 행 인덱스 기준

merge_df=pd.merge(df01,df02,right_index=True,left_index=True)
merge_df
```

```python
# join - default : outer join 
df01.join(df02)
df01.join(df02, how='inner')
```
