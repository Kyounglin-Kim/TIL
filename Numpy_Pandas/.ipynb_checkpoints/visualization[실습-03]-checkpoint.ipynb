```python
from glob import glob
```

```python
files=glob('./data/oil_data/지역*xls') # 파일에 대한 정보를 가져옴
#files
```

```python
# 파일 읽기 
tmp_raw=[]
for file in files:
    tmp=pd.read_excel(file,header=2)# 데이터가 3번째줄부터 있기때문 header 지정
    tmp_raw.append(tmp)
# tmp_raw
```

```python
# 리스트 -> 데이터프레임으로 만들기
oil_station = pd.concat(tmp_raw) # 리스트들을 결합..?
oil_station.head() # 인덱스가 중복되는 것을 확인됨
```

```python
oil_station.info() # 컬럼의 타입이 전부 object 
```

```python
# 휘발유만 대상으로 분석을 진행한다면
stations = pd.DataFrame({'oil_store' : oil_station['상호'],
                        'address' : oil_station['주소'],
                        'price' : oil_station['휘발유'],
                        'self_y_n' : oil_station['셀프여부'],
                        'company' : oil_station['상표']})

stations.head()
```

```python
# 파생변수를 생성
# 구별 주유 가격을 조사하기 위한 
stations['direct'] = [addr.split()[1] for addr in stations['address']]
stations.head()
```

```python
# 구별 정보를 확인하고 싶다면 - 특별시, 서울특별시 같은 이상한 값들이 존재함
stations['direct'].unique()
```

```python
# 서울특별시 -> 성동구 
# 특별시 -> 도봉구 변경하기
stations.loc[stations['direct']=='서울특별시','direct'] = '성동구'
stations.loc[stations['direct']=='특별시','direct'] = '도봉구'
stations['direct'].unique()
```

```python
# 가격정보 확인
stations[stations['price'] == '-']
```

```python
# 가격정보가 가입되지 않은 주요소는 제거
# 컬럼의 타입을 숫자형으로 변경

stations=stations[stations['price'] != '-']
stations.head()
```

```python
# 가격을 숫자형으로 변경
stations['price']=pd.to_numeric(stations['price'])

# 강사님 답 
#stations['price']=[int(price) for price in stations['price']]
```

```python
stations.info()
```

```python
stations.reset_index(inplace=True)
stations.head()
```

```python
# 필요없는 열 삭제
del stations['index']
# stations.drop('index',axis=1)
stations.head()
```

```python
stations.boxplot(column='price',by='self_y_n')
```

```python
# seaborn boxplot() 구현
sns.set_style('whitegrid')
plt.figure(figsize=(15,5))
sns.boxplot(x='company', y='price',hue='self_y_n',data=stations,palette='Set3')
sns.swarmplot(x='company', y='price',data=stations,palette='Set3')
plt.show()
```

```python
# 구별 주유 가격 확인 작업
import json
import folium
import googlemaps
```

```python
# 피봇 이용해서 구별 가격 정보로 변경하고 
direct_pivot=pd.pivot_table(stations,index=['direct'],values=['price'],aggfunc='mean')
direct_pivot
```

```python
# 구별 정보에 대해서 googlemaps 활용해서 시각화 해보기 
geo_str = json.load(open('./data/02. skorea_municipalities_geo_simple.json',encoding='utf-8'))
```

```python
map = folium.Map(location=[37.55, 126.98], zoom_start=10.5, tile='Stamen Terrain')

# Choropleth() : 구분된 경계를 표시할 때 사용하는 객체
folium.Choropleth(geo_data = geo_str,
            data = direct_pivot,
            columns = [direct_pivot.index,'price'],
            key_on='feature.id').add_to(map)
map
```

```python
# 주유가격 상위 10, 하위 10 주유소를 맵으로 시각화
oil_station_top10 = stations.sort_values(by='price', ascending=False).head(10)
display(oil_station_top10)
oil_station_bottom10 = stations.sort_values(by='price', ascending=True).head(10)
display(oil_station_bottom10)
```

```python
lat = []
lng = []
for n in oil_station_top10.index :
    tmp_addr =str(oil_station_top10['address'][n])
    tmp_map = gmaps.geocode(tmp_addr)
    # print(tmp_map)
    tmp_loc = tmp_map[0].get('geometry')
    lat.append(tmp_loc['location']['lat'])
    lng.append(tmp_loc['location']['lng'])

print(lat)
print(lng)

oil_station_top10['lat']=lat
oil_station_top10['lng']=lng
```

```python
lat = []
lng = []
for n in oil_station_bottom10.index :
    tmp_addr =str(oil_station_bottom10['address'][n])
    tmp_map = gmaps.geocode(tmp_addr)
    # print(tmp_map)
    tmp_loc = tmp_map[0].get('geometry')
    lat.append(tmp_loc['location']['lat'])
    lng.append(tmp_loc['location']['lng'])

print(lat)
print(lng)

oil_station_bottom10['lat']=lat
oil_station_bottom10['lng']=lng
```
