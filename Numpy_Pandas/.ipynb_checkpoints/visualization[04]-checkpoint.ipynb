- barplot -> 카테고리 값에 따른 실수 값의 평균과 편차를 표시하는 차트
- 평균은 막대의 높이 편차는 에러바

```python
# barplot 
# 스타일 테마를 설정 5가지 (darkgrid, whitegrid, dark, white, ticks)
sns.set_style('darkgrid')

# 그래프 객체를 생성하기 위한 피겨객체 얻어오기
fig = plt.figure(figsize=(15,5))

area01 = fig.add_subplot(1,3,1)
area01.set_title('titanic survived - sex')
area02 = fig.add_subplot(1,3,2)
area02.set_title('titanic survived - sex hug')
area03 = fig.add_subplot(1,3,3)
area03.set_title('titanic survived - sex dodge')

# 성별에 따른 생존률
sns.barplot(x='sex',y='survived',data=titanic,ax=area01)
# hug
sns.barplot(x='sex',y='survived',hue='class',data=titanic,ax=area02)
# dodge = False -> 누적 형식
sns.barplot(x='sex',y='survived',hue='class',dodge=False,data=titanic,ax=area03)

plt.show()


```

```python
# tips barplot()


# fig=plt.figure(figsize=(15,5))
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("요일별 전체 팁")
plt.show()
```

```python
# barplot 
# 스타일 테마를 설정 5가지 (darkgrid, whitegrid, dark, white, ticks)
sns.set_style('whitegrid')

# 그래프 객체를 생성하기 위한 피겨객체 얻어오기
fig = plt.figure(figsize=(15,5))

area01 = fig.add_subplot(1,3,1)
area01.set_title('titanic class')
area02 = fig.add_subplot(1,3,2)
area02.set_title('titanic class - who')
area03 = fig.add_subplot(1,3,3)
area03.set_title('titanic class - who(stacked)')

# 성별에 따른 생존률
sns.countplot(x='class',palette='Set1',data=titanic,ax=area01)
# hug
sns.countplot(x='class',hue='who',palette='Set2' ,data=titanic,ax=area02)
# # dodge = False -> 누적 형식
sns.countplot(x='class',hue='who',palette='Set3',dodge=False ,data=titanic,ax=area03)

plt.show()
```

```python
sns.countplot(x='day',palette='Set3',data=tips)
plt.title("요일별 팁의 빈도")
plt.show()
```

- stripplot() : scatter plot처럼 모든 데이터를 점으로 시각화 
- jitter=True 설정하면 위치를 무작위로 바꿔서 점들이 겹치지 않게 한다.

```python
sns.stripplot(x='day',y='total_bill',palette='Set2',data=tips,jitter=True)
plt.title("요일별 팁의 빈도")
plt.show()
```

```python
# 위치가 겹치면 좌우로 퍼치게 만든다
sns.swarmplot(x='day',y='total_bill',palette='Set2',data=tips) 
plt.title("요일별 팁의 빈도")
plt.show()
```

```python
# 타이나닉 데이터를 이용하여 서브플롯을 2개 영역으로 만들고
# class별 나이의 분포를 확인하라
# stripplot(),swarmplot()

# 스타일 테마를 설정 5가지 (darkgrid, whitegrid, dark, white, ticks)
sns.set_style('darkgrid')

# 그래프 객체를 생성하기 위한 피겨객체 얻어오기
fig = plt.figure(figsize=(15,5))

area01 = fig.add_subplot(1,2,1)
area01.set_title('titanic class - age distribution - strip')
area02 = fig.add_subplot(1,2,2)
area02.set_title('titanic class - age distribution - swarm')
# 나이에 따른 분포 
sns.stripplot(x='class',y='age',palette='Set3',data=titanic,jitter=True,ax=area01) 
# hug
sns.swarmplot(x='class',y='age',palette='Set2',data=titanic,ax=area02) 
plt.show()
```

```python
# 만약 tips에서 요일별, 성별 전체 팁 stripplot() 시각화

sns.stripplot(x='day',y='total_bill',palette='Set2',hue='sex',data=tips,jitter=True)
plt.title('요일별 전체 팁')
plt.legend(loc='best')
plt.show()
```

- boxplot(), violineplot()

```python
# 그래프 객체를 생성하기 위한 피겨객체 얻어오기

sns.set_style('whitegrid')

fig = plt.figure(figsize=(15,10))

area01 = fig.add_subplot(2,2,1)
area01.set_title('box')
area02 = fig.add_subplot(2,2,2)
area02.set_title('box')
area03 = fig.add_subplot(2,2,3)
area03.set_title('violin')
area04 = fig.add_subplot(2,2,4)
area04.set_title('violin')


sns.boxplot(x='alive', y='age',data=titanic,ax=area01)
sns.boxplot(x='alive', y='age',data=titanic,hue='sex',ax=area02)

sns.violinplot(x='alive', y='age',data=titanic,ax=area03)
sns.violinplot(x='alive', y='age',data=titanic,hue='sex',ax=area04)

plt.show()
```

- heatmap : 하나 이상의 카테고리 값의 변화를 색깔 변화로 시각화하는 차트

```python
# titanic
# 피벗테이블로 범주형 변수를 각각 행, 열로 재구분하여 정리
# 인덱스 성별, 열 클래스
table=titanic.pivot_table(index=['sex'],columns=['class'], aggfunc='size')
table

sns.heatmap(table,
           annot=True,fmt='d',cbar=True,
           linewidth=1) #fmt='d' 정수형으로 표현하겠다 cmap = rgb 코드 색
plt.show() # cbar=True False로 하면 오른쪽 옆 색깔 단위를 없앨수 있다
```

```python
# flights 연도, 월 별 승객수에 대한 heatmap
table2=flights.pivot('month','year','passengers')
table2

sns.heatmap(table2,
            cmap='YlGnBu',
           annot=True,fmt='d',cbar=True,
           linewidth=1) #fmt='d' 정수형으로 표현하겠다 cmap = rgb 코드 색
plt.show() # cbar=True False로 하면 오른쪽 옆 색깔 단위를 없앨수 있다
```
