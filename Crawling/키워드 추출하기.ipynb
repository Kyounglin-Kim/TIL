```python
k =[]
freq=[]
wordcount={}

for word in keyword_list:
    wordcount[word] = wordcount.get(word,0)+1
    keys = sorted(wordcount.keys())

for word in keys:
    print(word + ':' +str(wordcount[word]))
    k.append(word)
    freq.append(wordcount[word])
```

```python
keyword_freq = pd.DataFrame({'keyword' : k,
                            'freq' : freq})
```

### 도서관 키워드 빈도표 만들기

```python
# 단어 리스트 만들기
    

def word_list(book_df):
    
    keyword_list = []

    for i in range(len(book_df)):

        for a in range(len(book_df['keyword'][i].split('\''))):
            if a % 2 == 1:
                keyword_list.append(book_df['keyword'][i].split('\'')[a])
                
    return keyword_list
```

```python
# 단어 빈도표 반들기 

def word_freq(keyword_list):

    k =[]
    freq=[]
    wordcount={}

    for word in keyword_list:
        wordcount[word] = wordcount.get(word,0)+1
        keys = sorted(wordcount.keys())

    for word in keys:
        #print(word + ':' +str(wordcount[word]))
        k.append(word)
        freq.append(wordcount[word])
        
        return k,freq
```
