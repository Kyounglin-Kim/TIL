```python
import requests
from bs4 import BeautifulSoup
```

```python
from selenium import webdriver
import time
```

```python
title = []


for i in range(1,11):
    url = 'https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=MonthlyBest&BranchType=1&CID=0&page='+str(i)+'&cnt=1000&SortOrder=1'
    req = requests.get(url)
    content = req.content
    
    soup = BeautifulSoup(content,'html.parser')
    book_name_elems = soup.select('a.bo3 b')
   

    for item in range(len(book_name_elems)):
        title.append(book_name_elems[item].get_text())
```

```python
file_format = ['(지은이)','(옮긴이)','(엮은이)','(글쓴이)','(작가)','(글)','(그림)','(편곡)']
book_list=[]
check=[]
path = './driver/chromedriver.exe'
driver = webdriver.Chrome(path)

for b in range(1,11):
    url = 'https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=MonthlyBest&BranchType=1&CID=0&page='+str(b)+'&cnt=1000&SortOrder=1'
    driver.get(url)
    list = driver.find_elements_by_css_selector("div.ss_book_box > table > tbody > tr")
    
    for a in range(50):
        split_ls =list[a].text.split('\n')
        check.append(split_ls)        
```

```python
for c in range(len(check)):
    for i in check[c]:
        if (file_format[0] in i) | (file_format[1] in i) | (file_format[2] in i)|(file_format[3] in i)|(file_format[4] in i)|(file_format[5] in i)|(file_format[6] in i)|(file_format[7] in i):      
            book_list.append(i)
```
