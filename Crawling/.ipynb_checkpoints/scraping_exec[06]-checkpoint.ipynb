## 크롤링이 필요한지 크롤링 준비 & 이미지 스크래핑 & 자연어처리

- 이미지 스크랩핑

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from urllib.error   import HTTPError
from urllib.error   import URLError
from urllib.parse   import quote_plus
```

```python

base_url = 'https://search.naver.com/search.naver?where=image&sm=tab_jum&query='
keyword = input('검색어 입력 : ')
imgCnt = int(input('스크랩 할 이미지 개수  : '))

# 오픈을 할때 예외처리를 통해 만약 오류가 생기면 원인을 알수 있게 함수 있음

try :
    url = base_url + quote_plus(keyword) # quote_plus => 비 ASCII 텍스트를 ASCII 텍스트로 인코딩 --> 즉 한국어를 인식하기 위해 필요
    html = urlopen(url)  
except HTTPError as he :
    print('http error')
except URLError as ue :
    print('url error')
else :
    soup = BeautifulSoup(html.read(), 'html.parser',from_encoding='utf-8')
```

```python
img = soup.find_all('img', '_img')
```

```python
# img -> 리스트로 되어있다.
```

```python
cnt = 1
for i in img:
    imgUrl=i['data-source']
    with urlopen(imgUrl) as file: # url을 여는 작업
        with open('./images/'+ str(cnt)+'.jpg','wb') as imgFile:# 사진이름은 문자열이여야 하므로 str 사용
            img = file.read()
            imgFile.write(img)
            
    cnt +=1
    if cnt > imgCnt:
        break
        
    
```

- 검색키워드를 이용하여 원하는 제목과 링크만 가져오는 실습 
- csv 파일로 판다스의 DataFrame을 활용해서 저장해보자

```python
base_url = 'https://search.naver.com/search.naver?where=post&sm=tab_jum&query='
keyword = input('검색어 입력 : ')


# 오픈을 할때 예외처리를 통해 만약 오류가 생기면 원인을 알수 있게 함수 있음

try :
    url = base_url + quote_plus(keyword)
    html = urlopen(url)  
except HTTPError as he :
    print('http error')
except URLError as ue :
    print('url error')
else :
    soup = BeautifulSoup(html.read(), 'html.parser',from_encoding='utf-8')
```

```python
# sh_blog_title
titles = soup.find_all('a',{'class' : 'sh_blog_title'})
#titles
```

- 제목, 링크 가져와서 csv파일 만들기 -> 데이터프레임으로 만들기
- csv 읽기

```python
title_list=[]
link_list=[]


for tl in titles :
    title_list.append(tl.attrs['title']) # attrs 는 속성을 가져와야한다. text를 가져오는 것은 -> get_text()
    link_list.append(tl.attrs['href'])
  
```

```python
# pandas로 읽기

import pandas as pd
blog_df = pd.DataFrame({
    'title' : title_list,
    'link' : link_list
})

# csv 읽기

blog_df.to_csv('./data/blog_df.csv',  mode = 'w',encoding='utf-8',index=False) # csv데이터 생성
blog_df
```

```python
from time import sleep
from random import randint
import pandas as pd
from IPython.core.display import clear_output
from time import time
from bs4 import BeautifulSoup
from requests import get
```

```python
# loop 사이에 잘 진행되고 있는지를 보기 위한 indicator sentence
# 연습 단계
start_time = time()
# start_time
requests = 0

# 계속적으로 바뀌는 requests를 넣는데는 시간이 필요하다는 것을 확인하기 위한 단계
# n번 반복했을때 시간이 얼마나 걸리는건지
# frequencty = 반복 횟수를 걸린 시간으로 나눔
for _ in range(5) :
    requests += 1
    sleep(randint(1, 3))
    current_time = time()
    elapsed_time = current_time - start_time
    print('request : {}, frequency : {}'.format(requests, requests / elapsed_time))
    clear_output(wait = True)
```

```python
pages = [str(i) for i in range(1, 11)]
pages
```

```python
price_list =[]
title_list = []
link_list= []

# loop 사이에 잘 진행되고 있는지를 보기 위한 indicator sentence
start_time = time()
requests = 0

for page in pages :
    requests += 1
    url = get('https://search.shopping.naver.com/search/all?frm=NVSHATC&origQuery=%EA%B7%B8%EB%9E%98%ED%94%BD%EC%B9%B4%EB%93%9C&pagingIndex=' + page + '&pagingSize=40&productSet=total&query=%EA%B7%B8%EB%9E%98%ED%94%BD%EC%B9%B4%EB%93%9C&sort=rel&timestamp=&viewType=list')
    
    # request 에 delaytime 을 준다. -> 최대한 서버에서 사람이 행동하는 것 처럼 보여주기위해 delaytime을 둔다
    sleep(randint(1, 3)) # 랜덤하게 시간간격을 두게 하려고
    current_time = time()
    elapsed_time = current_time - start_time # elapsed는 현재시간-코드실행시간(?
    
    soup = BeautifulSoup(url.text, 'html.parser')
    
    # title, link, price
    # price 가져오기
    prices = soup.select('.price_num__2WUXn')
    price_text = [price.text for price in prices]
    price_list.extend(price_text)
    
    # title 가져오기
    titles = soup.select('.basicList_link__1MaTN')
    title_text = [title.text for title in titles]
    title_list.extend(title_text)
    
    link_text = [title['href'] for title in titles]
    link_list.extend(link_text)
```

```python
print(len(price_list))
print(len(title_list))
print(len(link_list))
```

```python
shop_df = pd.DataFrame({
    'title' : title_list,
    'price' : price_list,
    'link' : link_list
})

# csv 읽기

shop_df.to_csv('./data/shop_df.csv',  mode = 'w',encoding='utf-8',index=False) # csv데이터 생성
shop_df
```
